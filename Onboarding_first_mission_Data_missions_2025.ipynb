### 1.2 Clone the Repository and Conventional commits

Now that you've set up SSH and signing keys, you can clone a repository, create your branch and start working on it. Here are some good practices for commiting as Conventional Commits:

**Conventional Commits**

Conventional Commits is a specification for adding human and machine-readable meaning to commit messages. It standardizes the messages among all contributors, making the history easier to understand. It also allows for automated semantic versioning and release notes.

A conventional commit message should be structured as follows:
```
<type>(<scope>): <subject>
<BLANK LINE>
<body>
<BLANK LINE>
<footer>
```

- Type: This represents the nature of the changes. It must be one of the following:
- feat: A new feature
- fix: A bug fix
- docs: Documentation changes
- style: Changes that do not affect the meaning of the code (white-space, formatting, etc)
- refactor: A code change that neither fixes a bug nor adds a feature
- perf: A code change that improves performance
- test: Adding missing tests or correcting existing tests
- chore: Changes to the build process or auxiliary tools and libraries such as documentation generation

- Scope: This is optional and can be anything specifying the place of the commit change.

- Subject: A brief description of the changes, try to keep it under 50 characters.

- Body: A longer description of the changes. Not all commits are complex enough to warrant a body, so this is optional.

- Footer: Also optional, footer is the place to reference any issues or pull requests related to the commit.

Here's an example of a conventional commit:

```
feat(user-auth): add password reset functionality

Users can now reset their forgotten passwords. They will receive an email with a reset link, which will prompt them to enter a new password.

Resolves #123
```

By following the **Conventional Commits** standard, your team will have a more structured and readable commit history. It also helps in generating more accurate release notes and version numbers.

### example 2: Merge multiple tables on BQ



````python
df2 = client.query(
"""
with transactions as (
SELECT created_at transactions_created_at,
       status,
       amount,
       card_holder_name,
       card_number,
       merchant_id,
       capture_method
from maindb.transactions
)
,merchants as (
  SELECT * except(status), status as status_merchant
from maindb.merchants
)

select t.*,m.*
from transactions t
left join merchants m on m.user_id = t.merchant_id
where capture_method ='payment_link'
order by t.transactions_created_at DESC
limit 1000
"""
).to_dataframe()
````

### example 3: merge + calculating some features on BQ

````python
df3 = client.query(
"""
with transactions as (
SELECT created_at transactions_created_at,
       status,
       amount,
       card_holder_name,
       card_number,
       merchant_id,
       capture_method
from maindb.transactions
)
,merchants as (
  SELECT *
from maindb.merchants
)

,merchants_features as (
    select
      user_id,
      AVG(t.amount) as avg_amount,
      AVG(IF(t.status ='approved', t.amount, null)) as approved_avg_amount,
      AVG(IF(t.status ='approved' and t.capture_method='payment_link', t.amount, null)) as approved_pl_avg_amount,
      AVG(IF(t.status ='denied', t.amount, null)) as denied_avg_amount,
      # STDDEV() standard deviation
      # PERCENTILE_CONT() percentile
    from merchants m
    left join maindb.transactions t on t.merchant_id = m.user_id
    group by user_id
)

select t.*,mf.*,m.*
from transactions t
left join merchants m on m.user_id = t.merchant_id
left join merchants_features mf on mf.user_id = t.merchant_id
where capture_method ='payment_link'
order by t.transactions_created_at DESC
limit 1000
"""
).to_dataframe()
````

### example 3.1: merge + calculating some features on BQ  (cardholder)

````python
df4 = client.query(
"""
with transactions as (
SELECT created_at transactions_created_at,
       status,
       amount,
       card_holder_name,
       card_number,
       merchant_id,
       capture_method,
       cardholder_id,
       id
from maindb.transactions
where capture_method = 'payment_link'
)
, cardholders as (
  SELECT * from maindb.cardholders
)
,cardholder_features as (
  select
    t.cardholder_id,
    AVG(t.amount) AS avg_amount_cardholder,
    AVG(IF(t.status ='approved', t.amount, NULL)) AS approved_avg_amount_cardholder,
    AVG(IF(t.status ='denied', t.amount, NULL))   AS denied_avg_amount_cardholder,
    STDDEV(t.amount) as std_amount_cardholder
  from cardholders c
  left join transactions t on t.cardholder_id = c.id
  group by cardholder_id
)

select t.*, c.*, cf.*
from transactions t
left join cardholders c on t.cardholder_id=c.id
left join cardholder_features cf on t.cardholder_id=cf.cardholder_id
LIMIT 1000
"""
).to_dataframe()
````

### example 4: Function

Defining a custom function with create function and using external functions such as levenshtein that are actually .js functions we created! (you could create your own javascript functions in the future!)


vamos comparar o nome do carto na transao com o merchant_name (nome da store)

````python
leven = client.query(
"""
    CREATE TEMP FUNCTION normalized_text(x STRING)
    RETURNS STRING
    AS (LOWER(translate(x, "", "SZszYAAAAAACEEEEIIIIDNOOOOOUUUUYaaaaaaceeeeiiiidnooooouuuuyy")));

    -- select normalized_text('Alex');
    with merchants as (
      select
        user_id, merchant_name
      from maindb.merchants
    ),
    transactions as (
      select
        id, merchant_id, amount, card_holder_name
      from maindb.transactions
    )
select
    card_holder_name,
    merchant_name,
    `infinitepay-production`.udfs.levenshtein(normalized_text(m.merchant_name), normalized_text(t.card_holder_name)) lev_score
from transactions t
left join merchants m on t.merchant_id = m.user_id
order by lev_score desc
limit 1000

"""
).to_dataframe()
````


An array is an ordered list of zero or more elements of the same data type. You can create arrays in BigQuery by using the `ARRAY` keyword followed by `<data_type>[]`.

Here's an example of how to work with arrays:


```
sql
SELECT
name,
ARRAY<STRING>'apple', 'banana', 'cherry'] AS fruits
FROM
project.dataset.table
```

In this example, we're creating an array of strings with the names of fruits.

To access elements in an array, you can use the `OFFSET` keyword. Remember that arrays in BigQuery are zero-based.


```
sql
SELECT
name,
fruits[OFFSET(0)] AS first_fruit
FROM
project.dataset.table
```


**Structs**

A struct is a collection of fields, where each field has a name and a type. You can create structs in BigQuery by using the `STRUCT` keyword.

Here's an example of how to work with structs:
```
sql
SELECT
name,
STRUCT('apple' AS fruit, 10 AS quantity) AS fruit_info
FROM
project.dataset.table
```

In this example, we're creating a struct with two fields: `fruit` and `quantity`.

To access fields in a struct, you can use the dot notation.
```
sql
SELECT
name,
fruit_info.fruit,
fruit_info.quantity
FROM
project.dataset.table
```


In this example, we're accessing the `fruit` and `quantity` fields from the `fruit_info` struct.

Remember that you can also create arrays of structs and structs of arrays, allowing you to model complex data relationships.

### Example 6: Analytical functions

Similar to 2, but calculating for all data points with bigquery windowing

pandas -> groupby().rolling().sum()/count()

````python
client.query(
"""
with transactions as (
  select
    id,
    merchant_id,
    amount,
    status transaction_status,
    card_holder_name,
    card_number,
    created_at,
    unix_seconds(TIMESTAMP(created_at)) unix_seconds
  from maindb.transactions
  where created_at > '2021-09-16'
)

select
  t.*,
  SUM(if(t.transaction_status = 'approved', t.amount, NULL))
      OVER(PARTITION BY merchant_id ORDER BY unix_seconds RANGE BETWEEN 3600 PRECEDING AND CURRENT ROW ) sum_approved_amount_from_merchant_last_1h
from transactions t
where merchant_id = 2625 #cloudwalk
order by created_at DESC
limit 100
"""
).to_dataframe()
````

````python
from tyrell_feature_market.feature_market import FeatureMarket

fm = FeatureMarket()
left_query = f"""
    SELECT
        t.id as transaction_id,
        t.created_at as event_timestamp,
        merchant_id,
        SUBSTR(card_number, 0, 6) AS bin,
        CAST(m.cnae AS INT64) cnae,
        CAST(m.mcc AS INT64) mcc,
    FROM infinitepay-production.maindb.transactions t
    LEFT JOIN infinitepay-production.maindb.merchants m
    ON m.user_id = t.merchant_id
    LEFT JOIN infinitepay-production.maindb.cardholders ch
    ON ch.user_id = t.merchant_id
    WHERE capture_method='payment_link'
        AND t.created_at BETWEEN '2021-09-01' AND '2021-09-02'
"""
tables = ["bin_daily", "bin_attributes", "merchant_attributes"]
features = [
    "amount_chargebacked_pl_sum_from_merchant_last_30d_v3",
    "amount_chargebacked_pl_sum_from_merchant_last_90d_v3",
]
transformations = [
    {
        "method": "ratio",
        "features": [
            "amount_chargebacked_pl_sum_from_merchant_last_30d_v3",
            "amount_chargebacked_pl_sum_from_merchant_last_90d_v3",
        ],
    }
]

big_query_job = fm.get_historic_features(
    left_query, tables=tables, features=features, transformations=transformations, env='dev', model='onboarding'
)

df = big_query_job.to_df()

print(df.head())
````
